```python
import matplotlib.pyplot as plt # Python plot lib.

plt.plot([15.6, 14.2, 16.3, 18.2, 17.1, 20.2, 22.4]) # Plot value
plt.show()
```

```python
X = ["Mon", "Tue", "Wed", "Thur", "Fri", "Sat", "Sun"] # X-axis label
Y = [15.6, 14.2, 16.3, 18.2, 17.1, 20.2, 22.4] # Y-axis value, Y-axis label is automated mark

plt.plot(X, Y) # Define axis.
plt.xlabel("Day") # X-axis label naming.
plt.ylabel("Temp") # Y-axis label naming.
plt.show()
```

```python
X = ["Mon", "Tue", "Wed", "Thur", "Fri", "Sat", "Sun"]
Y1 = [15.6, 14.2, 16.3, 18.2, 17.1, 20.2, 22.4] # Double plot value - 1
Y2 = [20.1, 23.1, 23.8, 25.9, 23.4, 25.1, 26.3] # Double plot value - 2

plt.plot(X, Y1, label = "Seoul") # Plot lable
plt.plot(X, Y2, label = "Busan") # Plot label
plt.xlabel("Day")
plt.ylabel("Temp")

plt.legend(loc = "upper left") # Show what line's meaning
plt.title("Temperatures of Cities") # Showing title
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D

# get 3D axis
axis = plt.axes(projection='3d')

Z = np.linspace(0, 1, 100)
X = Z * np.sin(30 * Z)
Y = Z * np.cos(30 * Z)

axis.plot3D(X, Y, Z)
```
