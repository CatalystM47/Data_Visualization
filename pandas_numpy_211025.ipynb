```python
import numpy as np
import matplotlib.pyplot as plt

pure = np.linspace(1, 10, 100)
noise = np.random.normal(0, 1, 100)

signal = pure + noise

plt.plot(pure, pure, label='Tx signal')
plt.plot(pure, noise, label='noise')
plt.plot(pure, signal, label='Rx signal')

plt.xlabel("Time")
plt.ylabel("Signals")
plt.title("Review")
plt.legend(loc="upper left")

plt.show()
```

```python
import numpy as np
import time

# Normal python list performance
py_list = [i for i in range(10000000)]
start = time.time()
py_list = [i+2 for i in py_list]
end = time.time()
print("List = ", end - start)

# Numpy array performance
np_arr = np.array([i for i in range(10000000)])
start = time.time()
np_arr += 2
end = time.time()
print("Numpy = ", end - start)
```

```python
import numpy as np
import matplotlib.pyplot as plt

m, sigma = 10, 2
Y1 = np.random.randn(10000)
Y2 = m+sigma*np.random.randn(10000)

plt.figure(figsize=(10, 6))
plt.hist(Y1, bins = 20)
plt.hist(Y2, bins = 20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -2fi ~ +2fi 100 data create
X = np.linspace(-2 * np.pi, 2*np.pi, 100)

# apply sin func in numpy arr
Y1 = np.sin(X)
Y2 = 3 * np.sin(X)

plt.plot(X, Y1, X, Y2)
plt.show()
```
